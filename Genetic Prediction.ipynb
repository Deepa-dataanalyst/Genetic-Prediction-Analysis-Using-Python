# Genetic Prediction Analysis and Identifying Restricted Prediction Kits

```python
import pandas as pd
import random
```

```python
Title_Prediction=[
    "PeekaBoo Kit",
    "Happy Faces",
    "Little Feet Find",
    "Happy Lines",
    "Heritage Family Tree Kit",
    "Super Glue Relations",
    "Pink or Blue Kit",
    "MyDNa",
    "Find Me",
    "Heritage Therapy",
    "Bump and Baby",
    "Look at Me Kit",
    "Whisper Loud"]
Categories=[
    "Genetics",
    "Pregnancy",
    "Ancestry"]
```

```python
data=[]
for i in range(1,1001):
    data.append({
        "Prediction_Tool_ID":f"{i:03d}",
        "Title_Prediction":random.choice(Title_Prediction),
        "Category":
        random.choice(Categories),
        "Price":random.randint(3000,12000)})
    df=pd.DataFrame(data)
print(df.head())
print(df.shape)
```

```python


restricted_kits=[
    "PeekaBoo Kit",
    "Little Feet Find",
    "Pink or Blue Kit",
]
df["Restricted_Status"]=df["Title_Prediction"].apply(
    lambda x:"Restricted"
    if x in restricted_kits
    else "Permitted")
print(df)
```

### Total Kits

```python
print("Total Kits:",len(df))
```

### Restricted vs Non_Restricted

```python
print(df["Restricted_Status"].value_counts())
```

### Category Counts

```python
print(df["Category"].value_counts())
```

### Average Price by Category

```python
print("Average Price by",df.groupby("Category")["Price"].mean().round(2))
```

### Most Expensive Kit

```python
print(df.loc[df["Price"].idxmax()])
```

### Cheapest Kit

```python
print(df.loc[df["Price"].idxmin()])
```

### Show Only Restricted Kits

```python
restricted_kits=df[df["Restricted_Status"]=='Restricted']
print(restricted_kits)
```

### Percentage of Restricted Kits

```python
restricted_pct=(df["Restricted_Status"]=='Restricted').mean()
print("Percenatge of Restricted Kits is ",restricted_pct)
```
